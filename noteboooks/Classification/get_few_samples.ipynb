```python
import numpy as np
import json

disk_file = '../../datasets/classification/2002-2019_2/rebuttal_all_revised_filtered/all_samples_test.npy'
json_file = '../../datasets/classification/2002-2019_2/rebuttal_all_revised_filtered/dataset_filtered_test.json'

dataset = np.load(disk_file, allow_pickle=True).item()

with open(json_file, 'r') as f:
    json_content = json.load(f)
```

```python
import matplotlib.pyplot as plt

dataset_items = list(dataset.items())
print(len(dataset_items))
num_samples = len(dataset_items)

print(dataset_items[0][1].keys())
```

```python
# select 10 random samples
np.random.seed(2)
rand_idx = np.random.choice(num_samples, 20)

output_dir = './few_samples'
output_json = f'{output_dir}/few_samples.json'
output_json_dict = {}

for idx in rand_idx:


    img_key = dataset_items[idx][0]
    # print(dataset_items[idx][1].keys())

    img = dataset_items[idx][1]['image'] # size is 350x350


    json_entry = json_content[img_key]
    print(json_entry.keys())
    
    orig_img_file = img_key.split('_')[0]
    # classif = dataset_items[idx][1]['class1']+"_"+dataset_items[idx][1]['class2']+"_"+dataset_items[idx][1]['class3']
    classif = json_entry["McIntosh"]
    location_Lon = json_entry["centroid_Lon"]
    location_Lat = json_entry["centroid_Lat"]

    out_entry = {
        "orig_img_file": orig_img_file,
        "classif": classif,
        "location_Lon": location_Lon,
        "location_Lat": location_Lat
    }
    output_json_dict[img_key] = out_entry


    # get the central 256x256 crop
    cropped = img[47:303, 47:303]
    img = cropped

    # dump the image
    img_file = f'{output_dir}/{img_key}_{classif}.png'
    plt.imsave(img_file, img*255, cmap='gray', vmin=0, vmax=255)


    print(img.shape, img.min(), img.max(), img.dtype)

    fig, ax = plt.subplots(1, 1, figsize=(5, 5))
    ax.imshow(img, interpolation="none", cmap="gray", vmin=0, vmax=1)
    ax.set_title(img_key)
    ax.axis('off')
    plt.show()
    
    # break

with open(output_json, 'w') as f:
    json.dump(output_json_dict, f, indent=4)
```
